```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings = pd.read_csv('AB_NYC_2019.csv')
```

```python
listings
```

```python
#counts of the airbnb listing in each neighbourhood group of NYC
sn.countplot (x= 'neighbourhood_group', data = listings) #Crea la gráfica basado en la cuenta de hospedajes en cada barrio, especifico de donde viene la data
plt.show() #Muestra la gráfica de barras
```

```python
sn.barplot(x = 'neighbourhood_group', y= 'price', data = listings) #grafica que muestre el precio promedio de hospedaje en cada barrio
plt.show() #las lineas negras son intervalos de confianza
```

```python
sn.barplot(x = 'neighbourhood_group', y= 'price', data= listings, ci = False) #Sin los "confidence intervals" o intervalos de confianza
plt.show()
```
